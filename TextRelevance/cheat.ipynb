```python
from bs4 import BeautifulSoup
import re
import pandas as pd
from selenium import webdriver
import html
from time import sleep
from html.parser import HTMLParser
import random
import numpy as np
import requests
import selenium.common.exceptions as selexcept
from multiprocessing.dummy import Pool as ThreadPool
from multiprocessing import Lock, Value
from datetime import datetime
from selenium.webdriver.common.by import By
```

```python
qs={}
with open('queries.numerate.csv',encoding='utf-8') as f:
    for num_line, line in enumerate(f):
        data=line.strip().split('\t', 1)
        qs[int(data[0])]=(data[1])
urls={}
```

```python
def random_sleep(offset=1.5, length=4):
    sleep(random.random() * length + offset)
with webdriver.Edge() as driver:
    driver.get('https://mail.ru/')
    element = driver.find_element_by_xpath("//input[@id='q']")
    urls[1]=[]
    element = driver.find_element_by_xpath("//input[@id='q']")
    element.send_keys(qs[1])
    link = driver.find_element_by_id('search:submit')
    print("link found")
    link.click()
    random_sleep()
    items = driver.find_elements_by_class_name('light-link')
    for item in items:
        urls[1].append(item.get_attribute('data-url'))
    for q in qs:
        if q not in [52,315,356,395]:
            continue
        urls[q]=[]
        element = driver.find_element_by_xpath("//input[@id='q']")
        element.clear()
        element.send_keys(qs[q])
        link = driver.find_element_by_id('MSearch-submit')
        print("link found")
        link.click()
        random_sleep()
        items = driver.find_elements_by_class_name('light-link')
        for item in items:
            urls[q].append(item.get_attribute('data-url'))
```

```python
out= open("cheat_url.csv", 'w', encoding='utf-8')
for i in range(1,400):
    out.writelines(str(i)+'\t')
    for j in urls[i]:
        out.writelines((j.replace('http://','')).replace('https://','').replace('www.','')+' ')
    out.writelines('\n')
out.close()
```
